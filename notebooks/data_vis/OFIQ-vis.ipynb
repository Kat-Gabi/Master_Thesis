```python
def descriptive_plots_individual(data_path):
    # Load the dataset
    df = pd.read_csv(data_path, sep=';')

    # Set the aesthetic style of the plots
    sns.set_style("whitegrid")

    # Histograms for scalar scores with Seaborn
    scalar_columns = [col for col in df.columns if 'scalar' in col]
    for col in scalar_columns:
        plt.figure(figsize=(8, 4))
        sns.histplot(df[col], kde=True, bins=20, color='skyblue')
        plt.title(f'Histogram of {col}')
        plt.show()

    # scalar_columns = [col for col in df.columns if 'scalar' in col]
    # for col in scalar_columns:
    #     plt.figure(figsize=(8, 4))
    #     sns.histplot(df[col], kde=True, bins=20, color='skyblue')
    #     plt.title(f'Histogram of {col}')
    #     plt.show()

    # Improved Box plots for raw scores with Seaborn
    raw_columns = ['UnifiedQualityScore', 'BackgroundUniformity', 'IlluminationUniformity',
                   'LuminanceMean', 'Sharpness', 'CompressionArtifacts']
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df[raw_columns], palette="Set2")
    # sns.boxplot(data=df['scalar_columns'])
    plt.xticks(rotation=45)
    plt.title('Box Plots of Selected Raw Scores')
    plt.show()

    # Scatter plot for a pair of attributes (example: UnifiedQualityScore vs. Sharpness) with Seaborn
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='UnifiedQualityScore', y='Sharpness', data=df, hue='Sharpness', palette='coolwarm', size='Sharpness', sizes=(20, 200))
    plt.title('Scatter Plot of Unified Quality Score vs. Sharpness')
    plt.show()

    # Additional plots can be added based on specific needs or interests
```

```python
def boxplots_individual(data_path):
    # Load the dataset
    df = pd.read_csv(data_path, sep=';')

    # Set the aesthetic style of the plots
    sns.set_style("whitegrid")

    raw_columns = ['UnifiedQualityScore', 'BackgroundUniformity', 'IlluminationUniformity',
                   'LuminanceMean', 'Sharpness', 'CompressionArtifacts']
    for col in raw_columns:
        plt.figure(figsize=(8, 4))
        sns.boxplot(df[col], palette="Set2")
        plt.title(f'Boxplot of Raw scores of {col}')
        plt.show()
```
