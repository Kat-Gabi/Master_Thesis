## Canonical Age and Race Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import numpy as np
import cv2
```

```python
canonical = pd.read_csv('../../data/image_info_csvs/final_filtered_canonical_df_BIBLE.csv')
canonical["Type"] = "Canonical"

mixed = pd.read_csv('../../data/image_info_csvs/final_filtered_children_df_BIBEL.csv')
mixed["Type"] = "Mixed"
```

```python
can_mix = pd.concat([canonical, mixed], ignore_index=True)
```

## Racial distribution

```python
# Calculate the normalized counts (proportions)
count_df = can_mix.groupby(['Type', 'ethnicity']).size().reset_index(name='count')
total_counts = count_df.groupby('Type')['count'].transform('sum')
count_df['proportion'] = count_df['count'] / total_counts

# Create the Seaborn bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=count_df, x='ethnicity', y='proportion', hue='Type')
plt.title('Normalized Distribution of Ethnicity by Type')
plt.xlabel('Ethnicity')
plt.ylabel('Proportion')
plt.legend(title='Type')
plt.xticks(rotation=45)
plt.show()
```
